```python
import pandas as pd

dataset_fem = pd.read_csv('female_names.csv')
dataset_mas = pd.read_csv('male_names.csv')

```

```python

dataset_mas['name'] = dataset_mas['name'].astype(str).fillna('')
dataset_fem['name'] = dataset_fem['name'].astype(str).fillna('')
```

```python
def atributos(palabra):
    atributos = {}
    atributos['ultima letra'] = palabra[-1]
    atributos['primera letra'] = palabra[0]
    for letra in 'abcdefghijklmnopqrstuvwxyz':
        atributos["count({})".format(letra)] = palabra.lower().count(letra)
        atributos["has({})".format(letra)] = (letra in palabra.lower())
    return atributos
tagset = [(nombre, "mas") for nombre in dataset_mas["name"]] + [(nombre, "fem") for nombre in dataset_fem["name"]]
```

```python
import random
random.shuffle(tagset)
```

```python
tagset[:5]
```

```python
fset = [(atributos(nombres), g) for (nombres, g) in tagset]
train_tagset, test_tagset= fset[5000:], fset[:5000]

```

```python
print(fset[:5])
```

```python
import nltk
from nltk.classify import DecisionTreeClassifier, NaiveBayesClassifier
classifier_tree = DecisionTreeClassifier.train(train_tagset)
classifier_bayes = NaiveBayesClassifier.train(train_tagset)
```

```python
classifier_tree.classify(atributos("ana"))
```

```python
classifier_bayes.classify(atributos("joaquin"))
```

```python
print(nltk.classify.accuracy(classifier_bayes, test_tagset))
print(nltk.classify.accuracy(classifier_tree, test_tagset))
```
